## Streaming

Most models can stream their output content while it is being generated . By displaying output progressively, streaming significantly improves user experience, particularly for longer responses. calling stream() returns an interior that yields output chunks as they are produced . you can loop to process each chunk in real-time.


```python
model.invoke("Write a 200 words paragraph on generative ai")
```

```python
message = model.stream("Write a 200 words paragraph on generative ai")

for chunk in message:
    print(chunk.text, end="") # no extra newline per chunk
```

```python
for chunk in model.stream("What is the future of Ai taking over jobs in the world"):
    print(chunk.text, end="", flush=True)
```

## Batch

Batching is a collection of independent requests to a model can significantly improve performance and reduce costs , as the processing can be done in parallel.

```python
responses = model.batch([
    "why do parrots have colourful feathers?",
    "How do airplanes fly?",
    "what is quantum computing?"

])

for response in responses:
    print(response)
```
